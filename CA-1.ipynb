# Q1. Write a Python program to take details of first CA marks, second CA marks, third CA marks. Find out best two CA marks out of three CA.

```python
firstMarks = int(input("Enter 1st CA marks - "))
secondMarks = int(input("Enter 2nd CA marks - "))
thirdMarks = int(input("Enter 3rd CA marks - "))


if (firstMarks>secondMarks or firstMarks>thirdMarks) and (secondMarks>firstMarks or secondMarks>thirdMarks):
    print("Your best 2 CA marks are as follows:- ")
    print(firstMarks)
    print(secondMarks)
elif (secondMarks>firstMarks or secondMarks>thirdMarks) and (thirdMarks>firstMarks or thirdMarks>secondMarks):
    print("Your best 2 CA marks are as follows:- ")
    print(secondMarks)
    print(thirdMarks)
elif (thirdMarks>firstMarks or thirdMarks>secondMarks) and (firstMarks>secondMarks or firstMarks>thirdMarks):
    print("Your best 2 CA marks are as follows:- ")
    print(firstMarks)
    print(thirdMarks)
```
